```python
from simple_sf_cysch import *
import os
import numpy as np
```

```python
# Pull Salesforce data
stu_sec_df = get_sf_df('Student_Section__c', ['Id', 'Section__c', 'Local_Student_ID__c', 'Student__c', 'Student_Name__c', 'Active__c'])
stu_sec_df.rename(columns={'Id': 'Student_Section__c'}, inplace=True)
stu_sec_df = stu_sec_df[stu_sec_df['Active__c']==True]

section_df = get_sf_df('Section__c', ['Id', 'Name', 'School__c', 'Program__c', 'In_After_School__c'])
section_df.rename(columns={'Id': 'Section__c'}, inplace=True)
```

```python
df = section_df.merge(stu_sec_df, on='Section__c', how='outer')

df = df[df['Name'].str.contains('Tutoring')]

df['Name_cut'] = df['Name'].str[:-5]

df['Key'] = df['Student__c'] + df['Name_cut']

df = df.join(df.groupby(['Key'])['Section__c'].count(), on='Key', rsuffix='_count')

df = df[(df['Section__c_count']<2) & (df['In_After_School__c']=='In School')]
```
